# Visualizing attention for a single sentence with BERT

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from transformers import AutoTokenizer, AutoModel, utils
from bertviz import head_view, model_view

utils.logging.set_verbosity_error()

sentence_a = "While going to play near the riverbank, I met Raj, who was running an errand for his father to withdraw money from the bank."
model_name = "sentence-transformers/all-MiniLM-L6-v2" #"bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Important for recent transformers/torch stacks:
# force the eager attention implementation so attention tensors are actually materialized.
model = AutoModel.from_pretrained(
    model_name,
    attn_implementation="eager"
)
model.eval()

inputs = tokenizer(
    sentence_a,
    return_tensors="pt",
    truncation=True,
    add_special_tokens=True
)

with torch.no_grad():
    outputs = model(**inputs, output_attentions=True, return_dict=True)

attentions = outputs.attentions
input_ids = inputs["input_ids"][0]
tokens = tokenizer.convert_ids_to_tokens(input_ids)

print("Tokens:")
print(tokens)
print()

print("Number of layers:", len(attentions))
print("Attention shape of first layer:", tuple(attentions[0].shape))
```

## Interactive BertViz views

- `head_view(...)` lets you inspect specific heads
- `model_view(...)` gives a model-wide summary across layers and heads

If the widget does not render in your environment, the static matplotlib heatmaps above will still work.
